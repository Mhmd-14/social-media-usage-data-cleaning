```python
#Create a list containing some categories of socail media posts we can see in real life
list_category = ['Sport','Politcs','Fashion']

data_diction = {"Date":[],"Category":[],"Likes":[]}

for i in pd.date_range('2019-01-01',periods=10):
    for j in list_category:
        data_diction['Date'].append(i)
        data_diction['Category'].append(j)
        data_diction['Likes'].append(np.random.randint(0, 10000))
df1 = pd.DataFrame(data_diction)
df1
```

```python
likes = df1.groupby(by = ['Date','Category']).sum().values
likes
```

```python
df_likes =pd.DataFrame(likes.reshape(10,3),index= range(1,11),columns= ['Fashion','Politics','Sport'])
```
